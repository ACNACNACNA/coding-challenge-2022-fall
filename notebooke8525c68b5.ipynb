```python
def rSquared(xaxis, yaxis):
    sum = 0
    sumFromAvg = 0
    yavg = 0
    
    for value in yaxis:
        yavg += value
    yavg = yavg/len(yaxis)
    
    for val in range(len(xaxis)):
        difference = yaxis[val-1] - xaxis[val-1]
        sum += difference**2
        differenceFromAvg = yaxis[val-1] - yavg
        sumFromAvg += yavg**2 
    return (1 - (sum/sumFromAvg))
        

def OtherColumnAvg(table):
    avgratings = []
    for row in table:
        sum = 0
        for x in row:
            sum += x
        avgratings.append(sum/4)
    return avgratings
        
table = []
for col in range(5):
    ratingData = pd.read_csv('../input/carsforsale/cars_raw.csv', usecols=["ComfortRating", "InteriorDesignRating", "PerformanceRating", "ValueForMoneyRating", "ExteriorStylingRating", "ReliabilityRating"])
    ratingData = ratingData.values.tolist()
    testdata = ratingData
    print(ratingData[1])
    valueForMoney = []
    var = []
    
    for row in testdata:
        varHolder = row.pop(3)
        valueForMoney.append(varHolder)
        varHolder = row.pop(col)
        var.append(varHolder)
        
    variableCorrelation = rSquared(valueForMoney, var)
    otherRatings = OtherColumnAvg(testdata)
    otherRatingsCorrelation = rSquared(valueForMoney, otherRatings)
    table.append([variableCorrelation, otherRatingsCorrelation])
    
    print(table)
    


            
```

```python
newtable = []
rowOne = []
rowTwo = []

for row in table:
    row[0] = round(row[0],3)
    row[1] = round(row[1],3)
    rowOne.append(row[0])
    rowTwo.append(row[1])
    
newtable.append(rowOne)
newtable.append(rowTwo)
print(newtable)

        
```

```python
bars = ['Comfort', 'Interior Design', 'Performance', 'Exterior Styling', 'Reliability']
x = np.arange(len(bars))
fig, Finalgraph = plt.subplots()
rOneVar = Finalgraph.bar(x-0.15, newtable[0], 0.3, label = "R^2 Value of Rating vs. Value For Money")
rTwoOther = Finalgraph.bar(x+0.15, newtable[1], 0.3, label = "R^2 Value of Other Variable Ratings vs. Value For Money")
Finalgraph.set_xticks(x, bars)
Finalgraph.legend()
Finalgraph.bar_label(rOneVar, padding = 6)
Finalgraph.bar_label(rTwoOther, padding = 6)
plt.show()

```
